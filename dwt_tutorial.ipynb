```python
import numpy as np
import pywt
import matplotlib.pyplot as plt
from scipy.fftpack import fft
```

```python
# wavelet function and scaling finction
wave4 = np.array([(1 - np.sqrt(3)) / (4 * np.sqrt(2)),
                  -(3 - np.sqrt(3))/(4 * np.sqrt(2)),
                  (3 + np.sqrt(3))/(4 * np.sqrt(2)),
                  -(1 + np.sqrt(3))/(4 * np.sqrt(2))])
scal4 = np.array([(1 + np.sqrt(3)) / (4 * np.sqrt(2)),
                  (np.sqrt(3) + 3) / (4 * np.sqrt(2)),
                  (3 - np.sqrt(3)) / (4 * np.sqrt(2)),
                  (1 - np.sqrt(3)) / (4 * np.sqrt(2))])
```

```python
plt.rcParams["figure.figsize"] = (15, 6)
plt.figure()
plt.subplot(2,1,1)
plt.stem(wave4, label="wavelet function")
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
plt.subplot(2,1,2)
plt.stem(scal4,label="scaling function")
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
plt.figure()
plt.plot(np.abs(fft(wave4)), label="high-pass filter")
plt.plot(np.abs(fft(scal4)), label="low-pass filter")
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
```

```python
# signal
fs = 1024
t = np.arange(0, 1.0, 1.0 / fs)
features = [100,200,300]
intervals = [np.array([tt for tt in t]),
             np.array([tt if (0.6 < tt < 0.8) else 0 for tt in t]),
             np.array([tt if (0.2 < tt < 0.3) else 0 for tt in t]),
            ]
data = np.zeros(shape = (len(t),))
for i in range(len(features)):
    data += np.sin(2 * np.pi * features[i] * intervals[i])
plt.rcParams["figure.figsize"] = (15, 6)
plt.figure()
plt.plot(t,data)
plt.title("spectrum")
plt.xlabel("time")
plt.ylabel("amplitude")
```

```python
wavelet = pywt.Wavelet('db4')
levels = pywt.dwt_max_level(len(data), wavelet.dec_len)
coeffs = dict()
a = data
for level in range(1, levels + 1):
    (cA, cD) = pywt.dwt(a, wavelet)
    coeffs[level] = (cA, cD)
    a = cA
```

```python
for level, coeff in coeffs.items():
    n = len(coeff[0])
    plt.rcParams["figure.figsize"] = (15, 6)
    plt.figure()
    for i in range(1, 3):
        plt.subplot(2, 1, i)
        if i == 1:
            #plt.title("approximation component " + str(level), y = 1.08)
            plt.plot(np.linspace(0, t[-1], n), coeff[0], label = "approximation component " + str(level))
        else:
            #plt.title("detail component " + str(level), y = 1.08)
            plt.plot(np.linspace(0, t[-1], n), coeff[1], label = "detail component " + str(level))
        plt.xlabel("time")
        plt.ylabel("amplitude")
        plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
```
